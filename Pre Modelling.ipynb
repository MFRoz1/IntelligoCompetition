## Handling missing values

## Handling Outlier

## Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#Mendefinisikan nilai unique dari masing masing kolom yang akan di encodimg
col = sorted(data_new_drop['customer_city'].unique().tolist()) + sorted(data_new_drop['customer_state'].unique().tolist()) + sorted(data_new_drop['order_status'].unique().tolist())
print(col)
```

```python
# Mengubah kolom menggunakan OneHotEncoder
ohe = OneHotEncoder(handle_unknown = 'ignore')
encod = pd.DataFrame(ohe.fit_transform(data_new_drop[['customer_city', 'customer_state','order_status']]).toarray(), columns = col)
encod
```

# Data Pre Modelling

```python
data_new_drop.columns
```

```python
data_new_drop
```

```python
data_new_drop = data_new_drop.reset_index(drop=True)
data_new_drop
```

```python
data_final = data_new_drop.join(encod)
data_final
```

# Modelling

```python
# Melakukan pengecekan korelasi antar variabel menggunakan PCA
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

data_final_pca = pca.fit_transform(data_final)

explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2)
kmeans.fit(data_final_pca)
```
